```python
######## For Hackintosh with ryzen processor ####
!export MKL_DEBUG_CPU_TYPE=5
#################################################
from cv2_plt_imshow import cv2_plt_imshow
import cv2
```

```python
image = cv2.imread("faces/obama.jpg")
```

```python
cv2_plt_imshow(image)
```

```python
cv2.imwrite('obama.jpg', image)
```

```python
grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
HSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
cv2_plt_imshow(grey)
```

```python
cv2_plt_imshow(HSV)
```

```python
img = cv2.resize(image, (200, 200))
cv2_plt_imshow(img)
```

```python
row, col = image.shape[:2]
M = cv2.getRotationMatrix2D((col/2, row/2), 180,1)
```

```python
dst = cv2.warpAffine(image, M, (col, row))
```

```python
cv2_plt_imshow(dst)
```

```python
import numpy as np
```

```python
M = np.float32([[1, 0, -100], [0, 1 , -100]])
```

```python
dst = cv2.warpAffine(image, M, (col, row))
cv2_plt_imshow(dst)
```

```python
edge = cv2.Canny(image, 100, 500)
cv2_plt_imshow(edge)
```
